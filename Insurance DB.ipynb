```python
import pandas as pd
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head(10)
```

### Gráfico de dispersión: Prima vs. BMI

```python
# bmi és el valor de Body Mass Index i charges és l’import de risc de l’assegurança.
import seaborn as sns

sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

```python
# esbrinem si hi ha una correlació entre els valors bmi i el preu de la pòlissa

sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])

""" Segons el gràfic, efectivament hi ha una correlació positiva. A mesura que el valor bmi d’una p
ersona augmenta, també creix el preu. """
```

### Relación Prima vs. Fumador

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])

"""Aquí veiem que, clarament, si la persona és fumadora, el preu de la seva assegurança serà més alt. """
```

```python
# volem veure si la relació entre bmi i charges és diferent, en funció de si una persona és fumadora o no

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)

"""Efectivament, veiem que, per a una persona fumadora, el preu de la seva pòlissa augmenta considerablement quan té un bmi més alt, 
mentre que les persones no fumadores tenen un augment molt més petit. """
```

```python
# buscamos relación entre el precio de la póliza y la condición de fumador
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])

""" Aquest gràfic ens ensenya que la gran majoria de persones no fumadores tenen un preu inferior al de les persones fumadores. """
```
